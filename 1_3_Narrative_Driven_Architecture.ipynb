```python
# Narrative-Driven Architecture Generator
# Colab Notebook — Core Engine

import random
import textwrap
import matplotlib.pyplot as plt
import numpy as np

# For nicer console printing
WRAP_WIDTH = 90

def wrap(text):
    return textwrap.fill(text, width=WRAP_WIDTH)
```

```python
# Input: thematic seed

default_themes = [
    "solitude", "porosity", "ritual", "threshold",
    "erosion", "belonging", "memory", "transition"
]

print("Suggested themes:", ", ".join(default_themes))
theme = input("\nEnter a theme (or press Enter for random): ").strip().lower()

if not theme:
    theme = random.choice(default_themes)
    print(f"\nNo theme entered — using random theme: {theme!r}")
else:
    print(f"\nUsing theme: {theme!r}")
```

```python
# Narrative layer: generates a short atmospheric narrative from the theme

def generate_narrative(theme: str) -> str:
    # Very lightweight, template-based narrative engine
    mood_options = [
        "quiet and suspended in time",
        "charged with unspoken tension",
        "softly dissolving at the edges",
        "held between approach and withdrawal",
        "woven from fragments of memory",
        "oscillating between exposure and concealment",
    ]
    light_options = [
        "Light arrives in narrow, deliberate strokes,",
        "Light seeps in from unexpected seams,",
        "Light hovers at the periphery,",
        "Light falls in slow, patient intervals,",
        "Light traces the edges of absence,",
    ]
    material_options = [
        "surfaces absorb more than they reflect.",
        "materials feel worn, familiar, and slightly unfinished.",
        "boundaries blur between solid and void.",
        "textures invite touch but resist full legibility.",
        "edges fray into gradients rather than lines.",
    ]
    body_options = [
        "Bodies move cautiously,",
        "A lone figure drifts,",
        "Groups gather then disperse,",
        "Someone lingers at the threshold,",
        "Footsteps echo softly,",
    ]
    motion_options = [
        "testing how far they can retreat without disappearing.",
        "searching for a place to pause without being seen.",
        "tracing habitual paths that feel almost ritualistic.",
        "hovering between staying and leaving.",
        "learning the rhythm of the space by repetition.",
    ]

    mood = random.choice(mood_options)
    light = random.choice(light_options)
    material = random.choice(material_options)
    body = random.choice(body_options)
    motion = random.choice(motion_options)

    narrative = (
        f"This space grows out of the theme of {theme}. "
        f"The atmosphere is {mood}. "
        f"{light} and {material} "
        f"{body} {motion}"
    )

    return narrative

narrative = generate_narrative(theme)
print("\n--- NARRATIVE ---\n")
print(wrap(narrative))
```

```python
# Spatial metaphor layer: turns theme + narrative into spatial metaphors

def generate_spatial_metaphors(theme: str, narrative: str, n: int = 4):
    base_metaphors = [
        "as a narrowing corridor that compresses sound and light",
        "as a suspended platform hovering between two conditions",
        "as a sequence of rooms that gradually withdraw from the city",
        "as a porous wall that filters encounters rather than blocking them",
        "as a courtyard that collects fragments of passing time",
        "as a stair that never fully reveals where it leads",
        "as a gradient from exposure to shelter",
        "as a threshold that stretches into a room-long passage",
    ]

    random.shuffle(base_metaphors)
    selected = base_metaphors[:n]

    metaphors = []
    for m in selected:
        spatial_intention = random.choice([
            "to choreograph how one arrives and departs",
            "to slow down movement into a reflective pace",
            "to invite lingering without demanding it",
            "to frame glimpses rather than full views",
            "to hold multiple emotional states at once",
        ])
        material_logic = random.choice([
            "using heavier materials at the edges and lighter ones at the core",
            "thickening walls where privacy intensifies",
            "introducing perforations where contact is desired",
            "layering translucent and opaque surfaces",
            "allowing structure to become part of the spatial script",
        ])
        sensory = random.choice([
            "Sound is muffled, then released.",
            "Shadows become as important as objects.",
            "Temperature shifts mark the passage between zones.",
            "Echoes soften as one moves inward.",
            "The smell of outside air lingers at the edges.",
        ])
        program = random.choice([
            "This could host a retreat, an observatory, or a place of waiting.",
            "It could become a small library, a hermitage, or a listening room.",
            "It might hold a sequence of benches, alcoves, and quiet landings.",
            "It could frame a path to water, sky, or a single tree.",
            "It might serve as a threshold between home and city.",
        ])

        metaphors.append({
            "statement": f"{theme.capitalize()} {m}.",
            "spatial_intention": spatial_intention,
            "material_logic": material_logic,
            "sensory": sensory,
            "program": program,
        })

    return metaphors

metaphors = generate_spatial_metaphors(theme, narrative)

print("\n--- SPATIAL METAPHORS ---\n")
for i, m in enumerate(metaphors, 1):
    print(f"[{i}] {m['statement']}")
    print("   Intention:", m["spatial_intention"])
    print("   Material:", m["material_logic"])
    print("   Sensory:", m["sensory"])
    print("   Program:", m["program"])
    print()
```

```python
# Diagrammatic logic layer: abstract data for diagrams

def generate_diagram_data(theme: str, metaphors):
    """
    Returns a dict with simple geometric data for:
    - axial diagram
    - massing blocks
    - light paths
    - circulation flow
    """
    np.random.seed(len(theme) + len(metaphors))  # deterministic-ish per run

    # Axial: a main axis with side branches
    axis_length = 10
    axis_x = np.linspace(0, axis_length, 100)
    axis_y = np.zeros_like(axis_x)

    branches = []
    for _ in range(3):
        bx = np.linspace(random.uniform(2, 8), random.uniform(2, 8), 20)
        by = np.linspace(0, random.uniform(-3, 3), 20)
        branches.append((bx, by))

    # Massing: random rectangles
    massing_blocks = []
    for _ in range(5):
        cx = random.uniform(0, 10)
        cy = random.uniform(0, 6)
        w = random.uniform(0.8, 2.0)
        h = random.uniform(0.8, 2.0)
        massing_blocks.append((cx, cy, w, h))

    # Light paths: diagonal rays
    light_paths = []
    for _ in range(4):
        x0 = random.uniform(0, 10)
        y0 = random.uniform(4, 6)
        x1 = x0 + random.uniform(-2, 2)
        y1 = y0 - random.uniform(2, 4)
        light_paths.append(((x0, x1), (y0, y1)))

    # Circulation: a meandering polyline
    t = np.linspace(0, 1, 40)
    circ_x = 1 + 8 * t
    circ_y = 1 + 2 * np.sin(2 * np.pi * t)

    return {
        "axis": (axis_x, axis_y, branches),
        "massing": massing_blocks,
        "light": light_paths,
        "circulation": (circ_x, circ_y),
    }

diagram_data = generate_diagram_data(theme, metaphors)
print("Diagram data generated.")
```

```python
# Visual diagram 1: axial structure + circulation

axis_x, axis_y, branches = diagram_data["axis"]
circ_x, circ_y = diagram_data["circulation"]

plt.figure(figsize=(8, 4))
plt.plot(axis_x, axis_y, color="black", linewidth=2, label="Main axis")

for bx, by in branches:
    plt.plot(bx, by, color="gray", linewidth=1)

plt.plot(circ_x, circ_y, color="red", linewidth=2, linestyle="--", label="Circulation")

plt.title(f"Axial + Circulation Diagram — Theme: {theme}")
plt.axis("equal")
plt.axis("off")
plt.legend(loc="upper right")
plt.show()
```

```python
# Visual diagram 2: massing blocks + light paths

massing_blocks = diagram_data["massing"]
light_paths = diagram_data["light"]

plt.figure(figsize=(8, 4))

# Massing
for cx, cy, w, h in massing_blocks:
    rect = plt.Rectangle(
        (cx - w / 2, cy - h / 2),
        w, h,
        edgecolor="black",
        facecolor="lightgray",
        alpha=0.8,
    )
    plt.gca().add_patch(rect)

# Light paths
for (x0, x1), (y0, y1) in light_paths:
    plt.plot([x0, x1], [y0, y1], color="gold", linewidth=2)
    plt.scatter([x0], [y0], color="gold", s=20)

plt.title(f"Massing + Light Paths — Theme: {theme}")
plt.axis("equal")
plt.axis("off")
plt.show()
```

```python
# Text-only conceptual diagrams (ASCII) for export into docs / slides

def ascii_gradient(width=30, levels=5, fill_char="#", empty_char="."):
    lines = []
    for i in range(levels):
        filled = int(width * (i + 1) / levels)
        line = fill_char * filled + empty_char * (width - filled)
        lines.append(line)
    return "\n".join(lines)

def ascii_threshold(width=30, height=7):
    lines = []
    for r in range(height):
        if r in (0, height - 1):
            lines.append("#" * width)
        elif r == height // 2:
            lines.append("#" * (width // 3) + " " * (width // 3) + "#" * (width // 3))
        else:
            lines.append("#" + " " * (width - 2) + "#")
    return "\n".join(lines)

print("\n--- ASCII DIAGRAM: GRADIENT (e.g., exposure → shelter) ---\n")
print(ascii_gradient())

print("\n--- ASCII DIAGRAM: STRETCHED THRESHOLD ---\n")
print(ascii_threshold())
```

```python
# Final design logic summary tying narrative → metaphor → diagram

def summarize_design_logic(theme, narrative, metaphors):
    summary_parts = []

    summary_parts.append(
        f"The project begins from the theme of {theme}, "
        "treated not as a word but as an emotional condition."
    )

    summary_parts.append(
        "The narrative establishes the atmosphere and behavioral script of the space:"
    )
    summary_parts.append(wrap(narrative))

    summary_parts.append(
        "From this, a set of spatial metaphors emerges, each translating the theme "
        "into a possible architectural stance:"
    )

    for i, m in enumerate(metaphors, 1):
        summary_parts.append(
            f"{i}. {m['statement']} "
            f"It aims {m['spatial_intention']}, "
            f"{m['material_logic']} "
            f"{m['sensory']} {m['program']}"
        )

    summary_parts.append(
        "Diagrammatically, the project is explored through a main axis with branching "
        "paths, a meandering circulation line, clustered massing blocks, and oblique "
        "light paths. Together, these diagrams test how the narrative can be held in "
        "plan, section, and movement."
    )

    summary_parts.append(
        "The result is not a fixed design, but a story-based logic that can be adapted "
        "to different sites, programs, and scales while remaining faithful to the "
        f"core condition of {theme}."
    )

    return "\n\n".join(summary_parts)

summary = summarize_design_logic(theme, narrative, metaphors)

print("\n=== DESIGN LOGIC SUMMARY ===\n")
print(summary)
```

```python
# Package outputs into a single dictionary (for saving / further tooling)

output_package = {
    "theme": theme,
    "narrative": narrative,
    "metaphors": metaphors,
    "summary": summary,
    # diagram_data is already a dict of numpy arrays / tuples
    "diagram_data": diagram_data,
}

print("Output package ready. You can now:")
print("- Save it as JSON (minus numpy arrays),")
print("- Or feed it into another notebook / UI layer.")
```
